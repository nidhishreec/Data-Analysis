# DATA CLEANING

1. Check Null Values In The Dataset

```python
data.isnull().sum()
```

Observation : Null values are present in reenue column

2. Remove columns that are not necessary

```python
data = data.drop(columns='Metascore')
```

```python
data.columns
```

3. Fill the missing values

```python
data['Revenue (Millions)'] = pd.to_numeric(data['Revenue (Millions)'], errors='coerce')
data['Revenue (Millions)'] = data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].mean())
```

4. Check For Duplicate Data

```python
data.duplicated().any()
```

# DATA TRANSFORMATION

```python
data.rename(columns={"Revenue (Millions)": "Revenue"}, inplace=True)

data.rename(columns={"Runtime (Minutes)": "Runtime"}, inplace=True)
```

Classify Movies Based on Ratings [Good,Better and Best]

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
data
```

# ANALYSIS USING PLOTS

1. In Which Year There Was The Highest Revenue?

```python
plt.figure(figsize=(6, 4))

c = sns.barplot(
    data=data,
    x='Year',
    y='Revenue',  
    errorbar=None
)
plt.title("Revenue By Year")
plt.tight_layout()
plt.show()
```

Observations: 
- Revenue generally fluctuated between 80 and 110 units from 2006 to 2014, with a peak in 2009 and 2012.
- There's a noticeable decline in revenue in 2015 and 2016.

2. In Which Year There Was The Highest Voting?

```python
plt.figure(figsize=(8, 4))
c = sns.barplot(data=data, x='Year', y='Votes', errorbar=None)
c.bar_label(c.containers[0], fmt='%d')
plt.title("Total Votes By Year")
plt.tight_layout()
plt.show()
```

Observations

- Total votes were highest in 2012, peaking at 285,226.
- There's a significant drop in total votes from 2013 onwards, reaching a low of 48,591 in 2016.

3. Display Number of Movies Per Year

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Year', data=data)
plt.title("Number of Movies Per Year")
plt.tight_layout()
plt.show()
```

Observations:

- The number of movies released per year has steadily increased from 2006 to 2016.
- There was a significant surge in movie releases in 2015 and 2016, with 2016 having the highest count at nearly 300 movies.


4. Top 10 Highest Rated Movie Titles

```python
sns.barplot(x='Rating', y='Title', data=top_10)
plt.title("Top 10 Highest Rated Movie Titles")
plt.tight_layout()
plt.show()
```

Observations:

- All the top 10 highest-rated movies have ratings above 8.
- "The Dark Knight" appears to be the highest-rated movie among this top 10 list.

5. Distribution of Movie Runtime

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Runtime'], bins=10, kde=True)
plt.title("Distribution of Movie Runtime")
plt.show()
```

Observations:

- Most movie runtimes are between 100 and 120 minutes.
- Fewer very short or very long movies.

6. Does Movie Rating affect the Revenue

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Rating', y='Revenue', data=data)
plt.title("Movie Rating vs Revenue")
plt.tight_layout()
plt.show()
```

Observations:

- As movie ratings increase, there tends to be a general upward trend in revenue, suggesting higher-rated movies often earn more.
- There's a wide spread of revenue for any given rating, indicating that high ratings don't guarantee extremely high revenue, and some lower-rated movies can still earn a fair amount.

7. Distribution of Rating

```python
sns.kdeplot(data['Rating'],shade=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Density')
plt.show()
```

Observations:

Most ratings fall between 6 and 8, with very few movies receiving ratings below 4 or above 9.

8. How movie is distributied by rating category

```python
data['rating_cat'].value_counts().plot.pie(
    autopct='%1.1f%%',
    figsize=(4, 4),
    title='Movie Distribution by Rating Category',
    ylabel=''
)
plt.show()
```

Observations:
- "Excellent" is the largest category, accounting for 44.5% of movies.
- "Average" movies are the least common, making up 18.4% of the distribution.
