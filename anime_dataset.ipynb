Rating – Content rating based on age group (e.g., PG-13, R+, G).

Rank – Overall ranking based on the score among all anime.

Scored By – Total number of users who rated it.

Members – Total number of users who added it to their anime list (watched, plan to watch, etc.).

# DATA TRANSFORMATION

1. Convertioon of datatype to the required columns

```python
data['Episodes'] = pd.to_numeric(data['Episodes'], errors='coerce').astype('Int64')
data['Rank'] = pd.to_numeric(data['Rank'], errors='coerce').astype('Int64')
data['Score'] = data['Score'].astype(float)
data['Scored By'] = data['Scored By'].astype(float)
data['Popularity'] = data['Popularity'].astype(int)
data['Favorites'] = data['Favorites'].astype(int)
data['Members'] = data['Members'].astype(int)
```

2. Renaming the columns

```python
data = data.rename(columns={
    'English name': 'English_Name',
    'Other name': 'Other_Name'})
```

3. Categorizing based on the number of episodes

```python
bins = [0, 12, 50, 100, float('inf')]
labels = ['Short (<=12)', 'Medium (13-50)', 'Long (51-100)', 'Very Long (>100)']
data['episode_category'] = pd.cut(data['Episodes'], bins=bins, labels=labels)
data[['Name', 'Episodes', 'episode_category']].head(10)
```

# VISUAL ANALYSIS

1. Anime Count by Type

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='Type', order=data['Type'].value_counts().index)
plt.title("Number of Anime by Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Observations:
- TV series are the most prevalent type of anime.
- Special anime types has a very low count, appearing to be less than 10.

2. Top 10 favorite anime

```python
top_fav = data.nlargest(10, 'Favorites')[['Name', 'Favorites']].set_index('Name')

plt.figure(figsize=(8,4))
top_fav.plot(kind='bar', color='red', legend=False)
plt.title("Top 10 Anime by Favorites")
plt.ylabel("Favorites")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Observations:
- "Neon Genesis Evangelion" is the most favorited anime.
- The popularity of anime drops off sharply after the top few.

3. Distribution of Anime Ratings

```python
rating_counts = data['Rating'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Ratings")
plt.show()
```

Observations :

The vast majority of content is rated PG-13, significantly outweighing all other rating categories combined.

4. Distribution of Episodes

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Episodes'], bins=30, kde=True, color='purple')
plt.title("Distribution of Episodes")
plt.xlabel("Episodes")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Observations :
- The vast majority of anime have a relatively low number of episodes, specifically fewer than 50. 
- Anime with a large number of episodes are very rare.

5. Score vs Rank 

```python
plt.figure(figsize=(10,4))
sns.histplot(x='Rank', y='Score', data=data)
plt.title("Score vs Rank")
plt.xlabel("Rank")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```

Observations:

This indicates that top-ranked anime tend to have high scores, and scores decline significantly for lower-ranked entries.
